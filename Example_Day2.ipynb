```python
#To convert your image into a pencil sketch(computer vision)
import cv2 #computer vision library in python
from google.colab.patches import cv2_imshow
#inputting the image
image = cv2.imread("/content/2381781.jpg")
```

```python
image_show = cv2_imshow(image)
```

```python
#converting this BGR image to grayscale(black and white) image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#IMAGE INVERSION( IMAGE PROCESSING)
inverted_image = 255-gray_image
```

```python
blurred = cv2.GaussianBlur(inverted_image,(21,21),0)
inverted_blurred = 255-blurred
pencil_sketch= cv2.divide(gray_image, inverted_blurred, scale=256.0)
```

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
